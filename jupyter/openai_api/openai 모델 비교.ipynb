### 어떤 모델을 선택해야 할까?

- **최고의 성능과 복잡한 작업:** `gpt-5`
- **대부분의 웹/앱 서비스(가성비):** `gpt-5-mini`
- **실시간 음성/이미지 상호작용:** `gpt-4o`
- **긴 문서 분석:** `gpt-4.1`
- **이미지 생성:** `dall-e-3`
- **음성 처리:** `whisper-1` (음성→텍스트), `tts-1` (텍스트→음성)
- **의미 기반 검색 기능 구현:** `text-embedding-3-small` (가성비) 또는 `text-embedding-3-large` (최고 성능)

```python
import os
import openai
# from openai import OpenAI
openai.__version__
```

```python
# os.getenv('OPENAI_API_KEY')
```

```python

# 환경 변수에서 API 키를 로드합니다.
# 이 코드를 실행하기 전에 터미널에서 'export OPENAI_API_KEY=여러분의API키'를 설정하세요.
# client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
client = openai.OpenAI()

def get_ai_response(model_name, prompt):
    """지정된 모델과 프롬프트를 사용하여 AI의 응답을 요청하는 함수"""
    try:
        response = client.chat.completions.create(
            model=model_name,
            messages=[{"role": "user", "content": prompt}]
        )
        return response.choices[0].message.content
    except Exception as e:
        return f"오류가 발생했습니다: {e}"

# 이제 이 함수를 재사용하여 여러 모델을 테스트합니다.
```

```python
# 실습 1 코드
prompt = "딥러닝에 대해 비전공자가 이해하기 쉽게 설명하고, 실생활에 적용될 수 있는 미래 기술 예시를 3가지 들어줘."

response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=[{"role": "user", "content": prompt}]
)
response
```

```python
answer = response.choices[0].message.content
answer
```

```python

print("🚀 GPT-5-nano (빠르고 가벼운 모델)의 답변:")
print("-" * 50)
print(get_ai_response("gpt-5-nano", answer))
print("\n" * 2)
```

```python
%%time 
response = client.chat.completions.create(
    model="gpt-5-nano",
    messages=[{"role": "user", "content": prompt}]
)
answer = response.choices[0].message.content

print("🧠 GPT-5 (가장 강력한 모델)의 답변:")
print("-" * 50)
print(answer)
```
