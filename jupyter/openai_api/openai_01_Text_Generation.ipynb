**ChatCompletion vs Completion**
  
* Chat Completions 형식은 단일 사용자 메시지를 사용하여 요청을 구성함으로써 완료 형식과 유사해질 수 있습니다. 예를 들어, 다음 완료 프롬프트를 사용하여 영어에서 프랑스어로 번역할 수 있습니다.
```python
    Translate the following English text to French: "{text}"
```
* 동일한 채팅 프롬프트는 다음과 같습니다.

```python
    [{"role": "user", 
      "content": 'Translate the following English text to French: "{text}"'}
      ]
```

마찬가지로, 완성 API는 입력 내용을 적절히 포맷하여 사용자와 도우미 간의 채팅을 시뮬레이션하는 데 사용할 수 있습니다 .

이러한 API의 차이점은 각각에서 사용 가능한 기본 모델입니다. 채팅 완료 API는 가장 유능한 모델( gpt-4o)과 가장 비용 효율적인 모델( gpt-4o-mini)에 대한 인터페이스입니다.
