## 4-2. 이미지 생성 (Vision)

**OpenAI 문서 (이미지 생성)** : https://platform.openai.com/docs/guides/image-generation?image-generation-model=gpt-image-1

* Completions API : https://platform.openai.com/docs/guides/completions (openai v1.x)

* 프롬프트 예제 : https://platform.openai.com/docs/examples 

* https://platform.openai.com/docs/guides/vision

```python
from openai import OpenAI

client = OpenAI()

response = client.responses.create(
    model="gpt-4.1-mini",
    input=[{
        "role": "user",
        "content": [
            {"type": "input_text", "text": "what's in this image? 한국어로 답해줘."},
            {
                "type": "input_image",
                "image_url": "https://cdn.pixabay.com/photo/2025/05/04/17/47/dog-9578735_1280.jpg",
            },
        ],
    }],
)

print(response.output_text)
```
